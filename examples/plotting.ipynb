# Plotting

This notebook shows examples how to plot basic maps using GeoPandas together with the Matplotlib library.


```python
import geopandas as gpd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
cities = gpd.read_file(gpd.datasets.get_path('naturalearth_cities'))
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
world.plot(facecolor='red', edgecolor='black');
```

```python
cities.plot(facecolor='red', edgecolor='black');
```
